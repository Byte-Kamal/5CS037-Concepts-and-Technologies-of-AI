```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
scores = [69, 48, 84, 58, 48, 73, 83, 48, 66, 58, 84, 66, 64, 71, 64, 66, 69, 66, 69, 71, 81, 71, 73, 69, 66, 66, 64, 58, 64, 69, 69]

# Frequency Table
score_series = pd.Series(scores)

# Frequency count and relative frequency
freq_table = score_series.value_counts().sort_index()
relative_freq = freq_table / len(scores)
cumulative_freq = freq_table.cumsum()

# Combine into a DataFrame
freq_df = pd.DataFrame({
    'Frequency': freq_table,
    'Relative Frequency': relative_freq,
    'Cumulative Frequency': cumulative_freq
})
print("Frequency Table with Relative and Cumulative Frequencies:\n", freq_df)

# Grouped Frequency Table using Square Root Method
n = len(scores)  # Number of data points
k = int(np.sqrt(n))  # Class intervals using square root method
class_width = (max(scores) - min(scores)) / k  # Calculate class width

# Determine the class intervals
bins = np.arange(min(scores), max(scores) + class_width, class_width)

# Group data into the intervals
grouped_freq_table = pd.cut(scores, bins=bins, right=False)
grouped_freq = grouped_freq_table.value_counts()

# Display the grouped frequency table
print("\nGrouped Frequency Table:\n", grouped_freq)
```

```python
# Data
ages = [32, 40, 38, 45, 50, 34, 42, 36, 48, 55,
        39, 41, 37, 44, 49, 43, 35, 52, 46, 40,
        33, 47, 39, 51, 38, 44, 42, 36, 53, 41,
        37, 45, 39, 54, 40, 38, 36, 43, 50, 42,
        38, 55, 39, 37, 49, 46, 44, 40, 35, 52]

# Step 1: Grouped Frequency Table using Square Root Method
n = len(ages)
k = int(np.sqrt(n))  # Class intervals using square root method
class_width = (max(ages) - min(ages)) / k  # Calculate class width

# Determine the class intervals
bins = np.arange(min(ages), max(ages) + class_width, class_width)

# Group data into the intervals
grouped_freq_table = pd.cut(ages, bins=bins, right=False)
grouped_freq = grouped_freq_table.value_counts()

# Display the grouped frequency table
print("Grouped Frequency Table:\n", grouped_freq)

# Step 2: Histogram
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=bins, edgecolor='black', color='lightblue')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age Range')
plt.ylabel('Frequency')
plt.show()
```

```python
# Data: Class intervals and frequencies
weights = [(0, 40, 5), (40, 50, 6), (50, 60, 8), (60, 70, 4), (70, 100, 2)]

# Calculate Density (Frequency / Class Width)
class_intervals = [f"{start}-{end}" for start, end, _ in weights]
frequencies = [freq for _, _, freq in weights]
class_widths = [end - start for start, end, _ in weights]
densities = [freq / width for freq, width in zip(frequencies, class_widths)]

# Plot the histogram using density
plt.figure(figsize=(10, 6))
plt.bar(class_intervals, densities, color='lightgreen', edgecolor='black')
plt.title('Histogram with Unequal Class Widths (Density)')
plt.xlabel('Weight Range (gms)')
plt.ylabel('Density')
plt.show()
```

```python
# Data: Digits of pi
pi_digits = "14159265358979323846264338327950288419716939937510"

# Count the frequency of each digit from 0 to 9
digit_counts = {str(i): pi_digits.count(str(i)) for i in range(10)}

# Convert to DataFrame for easy handling
digit_df = pd.DataFrame(list(digit_counts.items()), columns=["Digit", "Frequency"])

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.bar(digit_df["Digit"], digit_df["Frequency"], color='lightcoral', edgecolor='black')
plt.title('Frequency Distribution of Digits in ')
plt.xlabel('Digit')
plt.ylabel('Frequency')
plt.show()
```
